```python
'''
Using exported SQL table. Cleaning null values and rearranging columns to create one master 
dataframe. From there, writing an Excel file to be imported into Tableau for additional 
analyses and visualizations. 
'''

#import pandas library
import pandas as pd
fg = pd.read_csv('https://raw.githubusercontent.com/pizacd/DrLupo_FallguysAnalyses/master/lupo_fallguys.csv')
```

```python
#converting dates column from object to datetime type
fg.date_played = pd.to_datetime(fg.date_played)
fg.sample(5)
```

```python
fg.info() 
```

```python
fg.crown.value_counts() #crown column should be boolean, instead it's f/t
```

```python
#Replacing all win/crown columns with 1s or 0s
fg.crown = fg.crown.replace({'t':1,'f':0})
fg.round1_win = fg.round1_win.replace({'t':1,'f':0})
fg.round2_win = fg.round2_win.replace({'t':1,'f':0})
fg.round3_win = fg.round3_win.replace({'t':1,'f':0})
fg.round4_win = fg.round4_win.replace({'t':1,'f':0})
fg.round5_win = fg.round5_win.replace({'t':1,'f':0})
fg.round6_win = fg.round6_win.replace({'t':1,'f':0})
```

```python
fg.crown.value_counts()
```

```python
fg.info()
```

```python
#Creating six separate dataframes for each round
rd1=fg[['match_id','date_played','round1_name','round1_type','round1_win']]
rd2=fg[['match_id','date_played','round2_name','round2_type','round2_win']]
rd3=fg[['match_id','date_played','round3_name','round3_type','round3_win']]
rd4=fg[['match_id','date_played','round4_name','round4_type','round4_win']]
rd5=fg[['match_id','date_played','round5_name','round5_type','round5_win']]
rd6=fg[['match_id','date_played','round6_name','round6_type','round6_win']]
```

```python
rd1.head()
```

```python
#Creating round column of all ones
rd1.isnull().sum()
rd1.insert(2,'round',[1]*len(rd1))
```

```python
#Creating round column of all twos
#Removing any row that has a null value
rd2.isnull().sum()
rd2 = rd2.dropna(how = 'any')
rd2.insert(2,'round',[2]*len(rd2))
```

```python
#Repeated process for rounds 3-6
rd3.isnull().sum()
rd3 = rd3.dropna(how = 'any')
rd3.insert(2,'round',[3]*len(rd3))
```

```python
rd4.isnull().sum()
rd4 = rd4.dropna(how = 'any')
rd4.insert(2,'round',[4]*len(rd4))
```

```python
rd5.isnull().sum()
rd5 = rd5.dropna(how = 'any')
rd5.insert(2,'round',[5]*len(rd5))
```

```python
rd6.isnull().sum()
rd6 = rd6.dropna(how = 'any')
rd6.insert(2,'round',[6]*len(rd6))
```

```python
#Made new column names, and assigned all Round dataframes to these column names
fall_col = ['match_id', 'date_played','round', 'name', 'type', 'won']
rd1.columns= fall_col
rd2.columns = fall_col
rd3.columns = fall_col
rd4.columns = fall_col
rd5.columns = fall_col
rd6.columns = fall_col
```

```python
#Concatenate all round dataframes into one
fallguys = pd.concat([rd1,rd2,rd3,rd4,rd5,rd6])
```

```python
fallguys.sample(10)
```

```python
fallguys.loc[fallguys.type == 'Final'].sample(10) #Sample of rows with data on the final round
```

```python
fallguys.type = fallguys.type.replace({'Type':'Team','Teams':'Team',}) #Doing some data cleaning where type column contained errors
```

```python
fallguys.name.value_counts()
```

```python
#Exporting data to Excel to be used in Tableau
fallguys.to_excel('DrLupo_fallguys_analyses.xlsx',sheet_name = 'DrLupo')
```
